```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data
import pandas as pd
path = "/content/drive/MyDrive/Supply Chain Project/supply_chain_data.csv"
df = pd.read_csv(path)


# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.1)

print("Creating comprehensive visualizations...")

# Figure 1: Product Analysis
fig1, axes = plt.subplots(2, 2, figsize=(16, 12))
fig1.suptitle('Product Type Analysis', fontsize=16, fontweight='bold')

# Product type distribution
product_counts = df['Product type'].value_counts()
axes[0, 0].bar(product_counts.index, product_counts.values, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
axes[0, 0].set_title('Product Type Distribution', fontweight='bold')
axes[0, 0].set_xlabel('Product Type')
axes[0, 0].set_ylabel('Count')
axes[0, 0].grid(axis='y', alpha=0.3)

# Average price by product type
avg_price = df.groupby('Product type')['Price'].mean().sort_values(ascending=False)
axes[0, 1].barh(avg_price.index, avg_price.values, color='#95E1D3')
axes[0, 1].set_title('Average Price by Product Type', fontweight='bold')
axes[0, 1].set_xlabel('Average Price ($)')
axes[0, 1].grid(axis='x', alpha=0.3)

# Revenue by product type
revenue_by_product = df.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False)
axes[1, 0].bar(revenue_by_product.index, revenue_by_product.values, color=['#FFA07A', '#98D8C8', '#F7DC6F'])
axes[1, 0].set_title('Total Revenue by Product Type', fontweight='bold')
axes[1, 0].set_xlabel('Product Type')
axes[1, 0].set_ylabel('Revenue ($)')
axes[1, 0].ticklabel_format(style='plain', axis='y')
axes[1, 0].grid(axis='y', alpha=0.3)

# Products sold by type
products_sold = df.groupby('Product type')['Number of products sold'].sum().sort_values(ascending=False)
axes[1, 1].bar(products_sold.index, products_sold.values, color=['#E74C3C', '#3498DB', '#2ECC71'])
axes[1, 1].set_title('Total Products Sold by Type', fontweight='bold')
axes[1, 1].set_xlabel('Product Type')
axes[1, 1].set_ylabel('Units Sold')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('product_analysis.png', dpi=300, bbox_inches='tight')
print("✓ Saved: product_analysis.png")

# Figure 2: Supply Chain Metrics
fig2, axes = plt.subplots(2, 2, figsize=(16, 12))
fig2.suptitle('Supply Chain Performance Metrics', fontsize=16, fontweight='bold')

# Stock levels distribution
axes[0, 0].hist(df['Stock levels'], bins=30, color='#6C5CE7', edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Stock Levels Distribution', fontweight='bold')
axes[0, 0].set_xlabel('Stock Level')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(axis='y', alpha=0.3)

# Lead times distribution
axes[0, 1].hist(df['Lead times'], bins=30, color='#00B894', edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Lead Times Distribution', fontweight='bold')
axes[0, 1].set_xlabel('Lead Time (days)')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(axis='y', alpha=0.3)

# Order quantities distribution
axes[1, 0].hist(df['Order quantities'], bins=30, color='#FD79A8', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Order Quantities Distribution', fontweight='bold')
axes[1, 0].set_xlabel('Order Quantity')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(axis='y', alpha=0.3)

# Shipping times distribution
axes[1, 1].hist(df['Shipping times'], bins=15, color='#FDCB6E', edgecolor='black', alpha=0.7)
axes[1, 1].set_title('Shipping Times Distribution', fontweight='bold')
axes[1, 1].set_xlabel('Shipping Time (days)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('supply_chain_metrics.png', dpi=300, bbox_inches='tight')
print("✓ Saved: supply_chain_metrics.png")

# Figure 3: Location & Supplier Analysis
fig3, axes = plt.subplots(2, 2, figsize=(16, 12))
fig3.suptitle('Location & Supplier Analysis', fontsize=16, fontweight='bold')

# Revenue by location
revenue_by_location = df.groupby('Location')['Revenue generated'].sum().sort_values(ascending=False)
axes[0, 0].bar(revenue_by_location.index, revenue_by_location.values, color='#E17055')
axes[0, 0].set_title('Total Revenue by Location', fontweight='bold')
axes[0, 0].set_xlabel('Location')
axes[0, 0].set_ylabel('Revenue ($)')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(axis='y', alpha=0.3)

# Supplier performance
supplier_revenue = df.groupby('Supplier name')['Revenue generated'].sum().sort_values(ascending=False)
axes[0, 1].barh(supplier_revenue.index, supplier_revenue.values, color='#74B9FF')
axes[0, 1].set_title('Revenue by Supplier', fontweight='bold')
axes[0, 1].set_xlabel('Revenue ($)')
axes[0, 1].grid(axis='x', alpha=0.3)

# Average manufacturing cost by supplier
avg_mfg_cost = df.groupby('Supplier name')['Manufacturing costs'].mean().sort_values(ascending=False)
axes[1, 0].barh(avg_mfg_cost.index, avg_mfg_cost.values, color='#A29BFE')
axes[1, 0].set_title('Avg Manufacturing Cost by Supplier', fontweight='bold')
axes[1, 0].set_xlabel('Manufacturing Cost ($)')
axes[1, 0].grid(axis='x', alpha=0.3)

# Production volumes by location
prod_by_location = df.groupby('Location')['Production volumes'].sum().sort_values(ascending=False)
axes[1, 1].bar(prod_by_location.index, prod_by_location.values, color='#55EFC4')
axes[1, 1].set_title('Total Production Volume by Location', fontweight='bold')
axes[1, 1].set_xlabel('Location')
axes[1, 1].set_ylabel('Production Volume')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('location_supplier_analysis.png', dpi=300, bbox_inches='tight')
print("✓ Saved: location_supplier_analysis.png")

# Figure 4: Shipping & Transportation
fig4, axes = plt.subplots(2, 2, figsize=(16, 12))
fig4.suptitle('Shipping & Transportation Analysis', fontsize=16, fontweight='bold')

# Shipping carrier performance
carrier_counts = df['Shipping carriers'].value_counts()
axes[0, 0].pie(carrier_counts.values, labels=carrier_counts.index, autopct='%1.1f%%',
               colors=['#FF7675', '#74B9FF', '#55EFC4'], startangle=90)
axes[0, 0].set_title('Shipping Carrier Distribution', fontweight='bold')

# Average shipping cost by carrier
avg_ship_cost = df.groupby('Shipping carriers')['Shipping costs'].mean().sort_values(ascending=False)
axes[0, 1].bar(avg_ship_cost.index, avg_ship_cost.values, color='#DFE6E9')
axes[0, 1].set_title('Avg Shipping Cost by Carrier', fontweight='bold')
axes[0, 1].set_xlabel('Carrier')
axes[0, 1].set_ylabel('Average Cost ($)')
axes[0, 1].grid(axis='y', alpha=0.3)

# Transportation mode distribution
transport_counts = df['Transportation modes'].value_counts()
axes[1, 0].bar(transport_counts.index, transport_counts.values, color=['#0984E3', '#6C5CE7', '#FDCB6E', '#E17055'])
axes[1, 0].set_title('Transportation Mode Distribution', fontweight='bold')
axes[1, 0].set_xlabel('Transportation Mode')
axes[1, 0].set_ylabel('Count')
axes[1, 0].grid(axis='y', alpha=0.3)

# Routes distribution
route_counts = df['Routes'].value_counts()
axes[1, 1].barh(route_counts.index, route_counts.values, color='#FD79A8')
axes[1, 1].set_title('Route Distribution', fontweight='bold')
axes[1, 1].set_xlabel('Count')
axes[1, 1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('shipping_transportation.png', dpi=300, bbox_inches='tight')
print("✓ Saved: shipping_transportation.png")

# Figure 5: Quality Control
fig5, axes = plt.subplots(1, 2, figsize=(16, 6))
fig5.suptitle('Quality Control Analysis', fontsize=16, fontweight='bold')

# Inspection results
inspection_counts = df['Inspection results'].value_counts()
colors_inspection = ['#00B894', '#FDCB6E', '#D63031']
axes[0].pie(inspection_counts.values, labels=inspection_counts.index, autopct='%1.1f%%',
            colors=colors_inspection, startangle=90, explode=[0.05, 0.05, 0.05])
axes[0].set_title('Inspection Results Distribution', fontweight='bold')

# Defect rates distribution
axes[1].hist(df['Defect rates'], bins=30, color='#D63031', edgecolor='black', alpha=0.7)
axes[1].set_title('Defect Rates Distribution', fontweight='bold')
axes[1].set_xlabel('Defect Rate (%)')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('quality_control.png', dpi=300, bbox_inches='tight')
print("✓ Saved: quality_control.png")

# Figure 6: Customer Demographics
fig6, axes = plt.subplots(1, 2, figsize=(16, 6))
fig6.suptitle('Customer Demographics Analysis', fontsize=16, fontweight='bold')

# Demographics distribution
demo_counts = df['Customer demographics'].value_counts()
axes[0].bar(demo_counts.index, demo_counts.values, color=['#6C5CE7', '#00B894', '#FDCB6E', '#E17055'])
axes[0].set_title('Customer Demographics Distribution', fontweight='bold')
axes[0].set_xlabel('Demographics')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

# Revenue by demographics
revenue_by_demo = df.groupby('Customer demographics')['Revenue generated'].sum().sort_values(ascending=False)
axes[1].barh(revenue_by_demo.index, revenue_by_demo.values, color='#74B9FF')
axes[1].set_title('Revenue by Customer Demographics', fontweight='bold')
axes[1].set_xlabel('Revenue ($)')
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('customer_demographics.png', dpi=300, bbox_inches='tight')
print("✓ Saved: customer_demographics.png")

# Correlation Heatmap
fig7, ax = plt.subplots(figsize=(14, 10))
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8}, ax=ax)
ax.set_title('Correlation Heatmap - Numerical Features', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
print("✓ Saved: correlation_heatmap.png")

plt.show()

print("\n" + "=" * 70)
print("ALL VISUALIZATIONS COMPLETED SUCCESSFULLY!")
print("=" * 70)
```
